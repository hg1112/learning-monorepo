## The Core Problem
```
In a user-level A/B split:
  Control users see old bidding algorithm
  Treatment users see new (more aggressive) bidding

BUT: both compete in the SAME auction.
  Treatment wins more → control loses more.
  Control looks WORSE than it actually is.
  Treatment looks BETTER than it actually is.
  
Solution: split by time (switchback) or by inventory pool.
```
