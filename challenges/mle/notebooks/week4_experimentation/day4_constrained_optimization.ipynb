## Key Math
```
Maximize: Σ value_i * win_i(b_i)
Subject to: Σ cost_i * win_i(b_i) ≤ Budget

Lagrangian: L(b, λ) = Σ value_i * win_i(b_i) - λ*(Σ cost_i * win_i(b_i) - Budget)

Optimal bid: b*_i = value_i / (1 + λ)
Where λ ≥ 0 is the shadow price of the budget constraint
```
