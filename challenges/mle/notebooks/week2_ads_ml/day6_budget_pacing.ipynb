## Key Concept
```
Goal: spend exactly $B over T hours
Target spend rate: B/T per hour
PID controller adjusts bid multiplier based on:
  - P (proportional): current spend rate vs. target
  - I (integral): cumulative underspend/overspend
  - D (derivative): rate of change of spend rate
```
