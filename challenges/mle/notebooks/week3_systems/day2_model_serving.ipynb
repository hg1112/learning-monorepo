## Latency Budget for a Typical Ads Request
```
Total: 100ms
├── Feature lookup (online store): 5–10ms
├── Candidate retrieval: 10–20ms
├── Model inference (ranking): 20–40ms
├── Calibration + bid calc: 1–2ms
└── Network + overhead: 20–30ms
```
